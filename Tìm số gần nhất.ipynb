```
import array as a
import bisect as bis

n = int(input())

arr = a.array('l', map(int, input().split()))

k, x = map(int, input().split())

high = bis.bisect(arr, x)
low = high - 1

near_x = a.array('l')
for i in range(0, k):
    if (high >= len(arr)):
        near_x.append(arr[low])
        low -= 1
    elif (low < 0):
        near_x.append(arr[high])
        high += 1
    else:
        if (arr[high] - x < x - arr[low]):
            near_x.append(arr[high])
            high += 1
        else:
            near_x.append(arr[low])
            low -= 1

print(*sorted(near_x), sep=' ')
```
